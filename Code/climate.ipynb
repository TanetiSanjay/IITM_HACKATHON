```python
import pandas as pd
```

```python
data = pd.read_csv('Data/Mean_Temp_IMD_2017.csv')
```

```python
data = data[data['YEAR']>=1970]
```

```python
data.head()
```

```python
data = data.drop(data.columns[13:], axis=1)
```

```python
new_df = pd.DataFrame()

def x(df, data):
    x = []
    y = []

    for _, row in data.iterrows():
        year = row['YEAR']
        for month in data.columns[1:]:
            # Append the value for the corresponding month and year
            y.append(f"{month}-{str(year)}")
            x.append(row[month])

    df['Month-Year'] = y
    df['Temperature'] = x

    return df

new_df = x(new_df, data)
new_df.to_csv('temperature.csv')
```

```python
import matplotlib.pyplot as plt 
plt.plot(new_df['Temperature'])
```
