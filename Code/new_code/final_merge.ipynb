```python
import yfinance as yf
```

```python
def get_finance_data(ticker_symbol = None):
    if ticker_symbol is None:
        ticker_symbol = "^NSEI"
    else:
        ticker_symbol = ticker_symbol
    
    financial_data = yf.download(ticker_symbol, start='2000-01-01', end='2017-12-31', interval='1mo')

    return financial_data
```

```python
finance_Data = get_finance_data()
```
