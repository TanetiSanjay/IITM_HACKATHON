```python
from yahooquery import Ticker

# Replace 'AAPL' with the desired stock symbol
stock_symbol = 'sbi'

# Create a Ticker object
ticker = Ticker(stock_symbol)

# Get ESG data through the history attribute
esg_data = ticker.history(period='1d', interval='1d')

# Display the ESG data
print("ESG Data:")
print(esg_data)


```

```python
# Extract specific ESG metrics
esg_score = esg_data[stock_symbol]['esgScore']
environmental_score = esg_data[stock_symbol]['environmentScore']
social_score = esg_data[stock_symbol]['socialScore']
governance_score = esg_data[stock_symbol]['governanceScore']

print("ESG Score:", esg_score)
print("Environmental Score:", environmental_score)
print("Social Score:", social_score)
print("Governance Score:", governance_score)
```

```python
import yfinance as yf

stock_symbol = 'sbi'

# Get ESG data
esg_data = yf.Ticker(stock_symbol).sustainability

# Display the ESG data
print("ESG Data:")
print(esg_data)

# Extract specific ESG metrics
if stock_symbol in esg_data.index.get_level_values(0):
    esg_score = esg_data.loc[stock_symbol, 'esgScore']
    environmental_score = esg_data.loc[stock_symbol, 'environmentScore']
    social_score = esg_data.loc[stock_symbol, 'socialScore']
    
    print(f"\nESG Score: {esg_score}")
    print(f"Environmental Score: {environmental_score}")
    print(f"Social Score: {social_score}")
else:
    print(f"\nESG data not available for {stock_symbol}")
```
