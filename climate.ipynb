```python
import pandas as pd
```

```python
data=pd.read_csv("Sub_Division_IMD_2017.csv")
```

```python
data.columns
```

```python
data.columns
import numpy as np
df=data
```

```python

# Filter data for the years 1980 to 2017
filtered_df = df[(df['YEAR'] >= 1980) & (df['YEAR'] <= 2017)]

# Create a new DataFrame to store the average rainfall month-wise
avg_rainfall_df = pd.DataFrame()

# Group filtered data by 'YEAR' and calculate the mean value for each month
avg_rainfall_df['JAN'] = filtered_df.groupby('YEAR')['JAN'].mean()
avg_rainfall_df['FEB'] = filtered_df.groupby('YEAR')['FEB'].mean()
avg_rainfall_df['MAR'] = filtered_df.groupby('YEAR')['MAR'].mean()
avg_rainfall_df['APR'] = filtered_df.groupby('YEAR')['APR'].mean()
avg_rainfall_df['MAY'] = filtered_df.groupby('YEAR')['MAY'].mean()
avg_rainfall_df['JUN'] = filtered_df.groupby('YEAR')['JUN'].mean()
avg_rainfall_df['JUL'] = filtered_df.groupby('YEAR')['JUL'].mean()
avg_rainfall_df['AUG'] = filtered_df.groupby('YEAR')['AUG'].mean()
avg_rainfall_df['SEP'] = filtered_df.groupby('YEAR')['SEP'].mean()
avg_rainfall_df['OCT'] = filtered_df.groupby('YEAR')['OCT'].mean()
avg_rainfall_df['NOV'] = filtered_df.groupby('YEAR')['NOV'].mean()
avg_rainfall_df['DEC'] = filtered_df.groupby('YEAR')['DEC'].mean()

# Display the new DataFrame
avg_rainfall_df
```
